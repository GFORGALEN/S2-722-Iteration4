```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import skewness, col

def create_spark_session():
    spark = SparkSession.builder \
        .appName("PySpark Basic Analysis") \
        .getOrCreate()
    return spark

def read_csv(spark, filename):
    df = spark.read.csv(filename, header=True, inferSchema=True)
    return df

def basic_analysis(df):
    # Drop columns
    df = df.drop('iso3', 'country', 'hdicode', 'region')
    
    # Display
    df.show(40)  # Show top 40 rows
    
    # Describe
    df.describe().show(40)  # Show top 40 rows of describe
    
    # Data types
    print("Data Types:\n", df.dtypes)
    
    # Count NULLs
    for column in df.columns:
        print(f"{column} NULL count: {df.filter(df[column].isNull()).count()}")
    
    # Skewness
    for column in df.columns:
        df.select(skewness(df[column])).show(40)

if __name__ == '__main__':
    filename = "Alldata.csv"  
    spark = create_spark_session()
    data = read_csv(spark, filename)
    basic_analysis(data)
```
